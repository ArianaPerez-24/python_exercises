*Cálculo de la ordenada al origen b: A continuación, se calcula la ordenada al origen b utilizando la fórmula de la recta y = mx + b. Para calcular b, se resta la pendiente m multiplicada por x1 a y1:*

```python
def ecuacion_de_recta(punto1, punto2):
    x1, y1 = punto1
    x2, y2 = punto2


    if x1 == x2:
        print("La pendiente es infinita, la recta es vertical.")
        return

    m = (y2 - y1) / (x2 - x1)

    # Calcular la ordenada al origen (b)
    b = y1 - m * x1
```

*Impresión de la ecuación de la recta: Finalmente, se imprime la ecuación de la recta en función de m y b. Dependiendo del valor de b, se formatea la impresión para incluir el signo "+" o "-" antes de b para que la ecuación sea legible. La ecuación se imprime con dos decimales:*

```python
def ecuacion_de_recta(punto1, punto2):
    x1, y1 = punto1
    x2, y2 = punto2


    if x1 == x2:
        print("La pendiente es infinita, la recta es vertical.")
        return

    m = (y2 - y1) / (x2 - x1)

    # Calcular la ordenada al origen (b)
    b = y1 - m * x1

    # Imprimir la ecuación de la recta
    if b < 0:
        print(f"La ecuación de la recta es: y = {m:.2f}x - {abs(b):.2f}")
    else:
        print(f"La ecuación de la recta es: y = {m:.2f}x + {b:.2f}")
```
